```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_train_path = "Dataset/data_train.csv"

URM_all_dataframe = pd.read_csv(data_train_path)
print(URM_all_dataframe.head(10))
```

```python
import numpy as np
import scipy.sparse as sps
from Data_manager.split_functions.split_train_validation_random_holdout import split_train_in_two_percentage_global_sample
from Evaluation.Evaluator import EvaluatorHoldout



URM_all = sps.coo_matrix((URM_all_dataframe['data'], (URM_all_dataframe['user_id'], URM_all_dataframe['item_id'])))
URM_all = URM_all.tocsr()

URM_train_validation, URM_test = split_train_in_two_percentage_global_sample(URM_all, train_percentage = 0.8)
evaluator_test = EvaluatorHoldout(URM_test, cutoff_list=[10])
```

```python
from Recommenders.Neural.MultVAE_PyTorch_Recommender import MultVAERecommender_PyTorch_OptimizerMask


recommender = MultVAERecommender_PyTorch_OptimizerMask(URM_train_validation, use_gpu=False)
recommender.fit(anneal_cap= 0.1063276470142029, batch_size= 128, dropout=0.15215292346427248, epochs= 131, l2_reg= 0.0014502013635276703, learning_rate= 0.00017395673258925492, total_anneal_steps= 389179.15872038255, evaluator_object=evaluator_test, validation_every_n= 5, stop_on_validation= True, lower_validations_allowed= 1, validation_metric= "MAP")
```
